**冒泡排序： **最优时间复杂度o(n),最坏时间复杂度为o(n2)  稳定

```python
def bubble_sort(alist):
    lenth = len(alist)
    for i in range(lenth-1):
        count = 0
        for j in range(lenth-1-i):
            if alist[j] > alist[j+1]:
                alist[j],alist[j+1] = alist[j+1],alist[j]
                count += 1
        if count == 0:
            return alist
    return alist
bubble_sort([2,14,615,523,123,6,2,12,513,1532,123541])
# bubble_sort([1,2,3,4])
        
```

**选择排序： **【24,25,23,1】 最优时间复杂度o(n2),最坏时间复杂度为o(n2)  不稳定

思路：假设有个虚拟的序列[ ]，找到序列中最小的值（假设是20），与第一个交换并放到虚拟序列中，则虚拟的序列是[20]，剩余的子序列就是[24,25,23];再找到子序列中最小的是23放到虚拟序列中，虚拟序列为[1,23],子序列为[24,25]；以此类推。选择排序是从后面未排序的子序列中找到一个最小的放到前面已经排序的虚拟序列中

```python
def select_sort(alist):
    lenth = len(alist)
    
    for i in range(lenth-1):
        min_index = i
        for j in range(i+1,lenth):
            if alist[min_index] > alist[j]:
                min_index = j
        alist[i],alist[min_index] = alist[min_index],alist[i]
    return alist
select_sort([2,14,615,523,123,6,2,12,513,1532,123541])
```

**插入排序： **【93,54,77,12】 最优时间复杂度o(n),最坏时间复杂度为o(n2)  稳定

思路：假设有个虚拟的序列[93]，依次遍历剩余子序列[54,77,12]，54与虚拟序列中的93比较并加入虚拟序列中，则虚拟的序列是[54,93]，剩余的子序列就是[77,12];再遍历子序列中77并与虚拟序列中的54,93比较并加入序列中，则虚拟序列为[54,77，293],子序列为[12]；以此类推。插入排序与选择排序相反，插入排序是选择子序列的元素与前面虚拟序列中的元素比较

```python
def insert_sort(alist):
    lenth = len(alist)
    for i in range(1,lenth):
        j = i
        while j > 0:
            if alist[j] < alist[j-1]:
                alist[j],alist[j-1] = alist[j-1],alist[j]
                j -= 1
            else:
                break
    return alist
insert_sort([2,14,615,523,123,6,2,12,513,1532,123541])
```

**希尔排序： **是希尔排序的改进版，实质上是一种分组插入方法 不稳定

思路：希尔排序是把记录按下标的一定增量分组，对每组使用直接插入排序算法排序；随着增量逐渐减少，每组包含的关键词越来越多，当增量减至1时，整个文件恰被分成一组，算法便终止

专家们提倡，几乎任何排序工作在开始时都可以用希尔排序，若在实际使用中证明它不够快，再改成快速排序这样更高级的排序算法. 本质上讲，希尔排序算法是直接插入排序算法的一种改进，减少了其复制的次数，速度要快很多。 原因是，当文件初态基本有序时直接插入排序所需的比较和移动次数均较少。当n值很大时数据项每一趟排序需要移动的个数很少，但数据项的距离很长。当n值减小时每一趟需要移动的数据增多，此时已经接近于它们排序后的最终位置。 正是这两种情况的结合才使希尔排序效率比插入排序高很多。Shell算法的性能与所选取的分组长度序列有很大关系。只对特定的待排序记录序列，可以准确地估算关键词的比较次数和对象移动次数。想要弄清关键词比较次数和记录移动次数与增量选择之间的关系，并给出完整的数学分析，今仍然是数学难题。



```python
def shell_sort(alist):
    lenth = len(alist)
    gap = lenth // 2
    while gap >=1:
        for j in range(gap,lenth):
            i = j
            while i > 0:
                if alist[i] < alist[i-gap]:
                    alist[i], alist[i-gap] = alist[i-gap], alist[i]
                    i -= gap
                else:
                    break
        gap = gap // 2
    return alist
shell_sort([2,14,615,523,123,6,2,12,513,1532,123541])
```

**快速排序   重要需掌握 ： **是通过一趟排序将要排序的数据分割成独立的两部分，其中一部分的所有数据都比另外一部分的所有数据都要小，然后再按此方法对这两部分数据分别进行快速排序，整个排序过程可以递归进行，以此达到整个数据变成有序序列。 最坏时间复杂度为nlog2(n)（每次mid_value都是在中间），最优时间复杂度为o(n2)  不稳定



```python
def quick_sort(alist,first,last):
    if first >= last:
        return alist
    mid_value = alist[first]
    low = first
    high = last
    while low < high:
        # high左移
        while low < high and alist[high] >= mid_value :
            high -= 1
        alist[low] = alist[high]
        # low右移
        while low < high and alist[low] < alist[high]:
            low += 1
        alist[high] = alist[low]
    # 当循环退出时，low==high
    alist[high] = mid_value
    
    # 对high左边的列表执行快速排序
    quick_sort(alist, 0, low-1)
    # 对high左边的列表执行快速排序
    quick_sort(alist, low+1, last) # last表示原先的low-1
shell_sort([2,14,615,0,1532,123,6,2,12,6,513,1532,123541])
```

**归并排序： **是采用分治法的一个典型应用。归并排序的思想就是先递归分解数组，再合并数组 最坏时间复杂度为nlog2(n)，最优时间复杂度为nlog2(n)（分解过程不产生复杂度）,但是归并排序需要额外的空间开销  稳定



```python
def merger_sort(alist):
    n = len(alist)
    if n <= 1:
        return alist
    mid = n // 2
    # 采用归并排序后形成有序的**新的列表**
    left_alist = merger_sort(alist[:mid])
    right_alist = merger_sort(alist[mid:])
    left_pointer, right_pointer = 0, 0
    result = []
    while left_pointer < len(left_alist) and right_pointer < len(right_alist):
        if left_alist[left_pointer] <= right_alist[right_pointer]:
            result.append(left_alist[left_pointer])
            left_pointer += 1
        else:
            result.append(right_alist[right_pointer])
            right_pointer += 1
    result += left_alist[left_pointer:]
    result += right_alist[right_pointer:]
    return result
merger_sort([2,14,615,0,1532,123,6,2,12,6,513,1532,123541])
```
